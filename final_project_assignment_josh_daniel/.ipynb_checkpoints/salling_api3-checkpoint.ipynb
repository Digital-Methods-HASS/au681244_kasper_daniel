```python
from datetime import datetime

now = datetime.now() # Defining current time

current_time = now.strftime("%d-%m-%y") # Choosing date and time values with strftrime() and saving them

container = [] # Creating empty container for future request values

def query_function(query):
    endpoint_string_salling = "https://api.sallinggroup.com/" # Defining endpoint
    query_city = f"v1/food-waste/?zip={query}" # string function to choose city postal-code
    custom_auth_key = {'Authorization': 'Bearer 8d345a30-b653-4627-81f5-7778e7cd79a0'} # Auth ID from Salling Group API
    response_salling = requests.get(endpoint_string_salling + query_city, headers = custom_auth_key) # Creating the request
    response_salling.json() # Converting the data to json-type
    
    json_data = response_salling.json()[0] # Testing indexvalue
    json_container = json_data["clearances"] # Creating container with the json-string
    for iteration in json_container: # Iterating over the json-container
        for keys in iteration: # Iterating over each product in the container
            container.append([iteration[keys]]) # Appending the "offer" dict and the "product" dict seperately 

    offer_lists = container[::2] # loop over all "offer" dicts from each product (every second index)
    product_lists = container[1::2] # loop over all "product" dicts from each product
    
    # Making 2 seperate lists for "offer" and "product"
    
    final_offer = []
    for iteration in offer_lists:
        final_offer.extend(iteration)
    
    final_product = []
    for iteration in product_lists:
        final_product.extend(iteration)

    # Converting both lists to seperate dataframes    
        
    dataframe_product = pd.DataFrame(final_offer)
    dataframe_offer = pd.DataFrame(final_product)
    
    # Then merging them by their common index-values
    
    dataframe_foodwaste = pd.merge(dataframe_product, dataframe_offer, left_index=True, right_index=True)
    
    # We write our merged dataframe to a csv-file with postalcode and timestamp
    
    dataframe_foodwaste.to_csv(f"data/data_foodwaste_postalcode{query}_{current_time}.csv", index=False)

# We can now call our function with the desired zip-code to aquire csv-files    
    
query_function(6700)
```

# Pre-Processing
- the dataframes are very messy, especially the "category" column. 
- so we split the whole category string into its different parts in both languages (overall-type_of-specific)

### pre-process cph_files

```python
# split danish categories
df_cph[['category1','category2','category3']]= df_cph.categories.str.split(",",expand=True) # split by the two languages
df_cph["category1"]= df_cph['category1'].replace("{'da': '", '', regex=True).replace("'", "", regex=True) # replace unnecessary 

df_cph[['overordnet_kategori','type_kategori','specifik_kategori',"none_1","none_2", "none_3", "none_4", "none_5"]]= df_cph["category1"].str.split(">",expand=True) # split ">"

df_cph = df_cph.drop(columns=['none_1', 'none_2', "none_3", "none_4", "none_5", "category3"]) # drop all columns
#df_cph.head()

# split english categories
df_cph["category2"]= df_cph['category2'].replace("'en': '", '', regex=True).replace("'}", "", regex=True)# replace unnecessary 
df_cph[['overall_category','type_of','specific_category',"none_5","none_6","none_7","none_8", "none_9"]] =df_cph["category2"].str.split(">",expand=True) # split ">"


df_cph = df_cph.drop(columns=['none_5', 'none_6', "none_7", "none_8","none_9"]) # drop all columns
df_cph.sort_values(by='percentDiscount', ascending=False)

df_cph["postcode"] = 2000



```
