```python
from datetime import datetime

now = datetime.now() # Defining current time

current_time = now.strftime("%d-%m-%y") # Choosing date and time values with strftrime() and saving them

container = [] # Creating empty container for future request values

def query_function(query):
    endpoint_string_salling = "https://api.sallinggroup.com/" # Defining endpoint
    query_city = f"v1/food-waste/?zip={query}" # string function to choose city postal-code
    custom_auth_key = {'Authorization': 'Bearer 8d345a30-b653-4627-81f5-7778e7cd79a0'} # Auth ID from Salling Group API
    response_salling = requests.get(endpoint_string_salling + query_city, headers = custom_auth_key) # Creating the request
    response_salling.json() # Converting the data to json-type
    
    json_data = response_salling.json()[0] # Testing indexvalue
    json_container = json_data["clearances"] # Creating container with the json-string
    for iteration in json_container: # Iterating over the json-container
        for keys in iteration: # Iterating over each product in the container
            container.append([iteration[keys]]) # Appending the "offer" dict and the "product" dict seperately 

    offer_lists = container[::2] # loop over all "offer" dicts from each product (every second index)
    product_lists = container[1::2] # loop over all "product" dicts from each product
    
    # Making 2 final seperate lists for "offer" and "product"
    
    final_offer = []
    for iteration in offer_lists:
        final_offer.extend(iteration)
    
    final_product = []
    for iteration in product_lists:
        final_product.extend(iteration)

    # Converting both lists to dataframe    
        
    dataframe_product = pd.DataFrame(final_offer)
    dataframe_offer = pd.DataFrame(final_product)
    
    # Writing each list to a separate csv-file with a string-fuction to name them properly
    # Adela sagde at row_join kunne være en måde man kombinerede dataframes. Vil du kigge på det?
    
    dataframe_product.to_csv(f"data/data_product_postalcode{query}_{current_time}.csv", index=False)
    dataframe_offer.to_csv(f"data/data_offer_postalcode{query}_{current_time}.csv", index=False)

query_function(8000)
```
